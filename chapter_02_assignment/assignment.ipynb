```python
# import libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen

# Aapka share kiya hua HTML data
html_data = urlopen('https://www.pythonscraping.com/pages/page3.html')

soup = BeautifulSoup(html_data, 'html.parser')

# 1. Sabse pehle hum saari rows nikalenge jinki class "gift" hai
gift_rows = soup.find_all('tr', class_='gift')

# 2. Har row ke andar loop chalayenge
for row in gift_rows:
    # Row ke andar maujud saare <td> tags (cells) ki list bana li
    cells = row.find_all('td')
    
    # Index ke hisab se data alag kiya:
    # cells[0] = Title, cells[1] = Description, cells[2] = Cost, cells[3] = Image
    title = cells[0].text.strip()
    description = cells[1].text.strip()
    cost = cells[2].text.strip()
    
    # Image hamesha direct text nahi hoti, hume <img> tag ka 'src' attribute chahiye
    img_tag = cells[3].find('img')
    image_url = img_tag['src'] if img_tag else "No Image"
    
    # Data ko print karne ka treeqa
    print(f"🎁 Title: {title}")
    print(f"📝 Description: {description}")
    print(f"💰 Cost: {cost}")
    print(f"🖼️ Image URL: {image_url}")
    print("-" * 50) # Separation line
```

### By using find() ,find_all()  ,children parent ,previous_sibling ,get_text()

```python
# Import libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen


# --------------------------------------------------
# 1. Create BeautifulSoup object
# --------------------------------------------------

url = 'https://www.pythonscraping.com/pages/page3.html'

html = urlopen(url)
bs = BeautifulSoup(html.read(), 'html.parser')


# --------------------------------------------------
# 2. Find the specific table
# Using find()
# --------------------------------------------------

table = bs.find('table', {'id': 'giftList'})


# --------------------------------------------------
# 3. Get direct children of the table
# Using children
# --------------------------------------------------

for row in table.children:

    # Make sure the child is a <tr> tag
    if row.name != 'tr':
        continue

    # --------------------------------------------------
    # 4. Find all <td> elements inside the row
    # Using find_all()
    # --------------------------------------------------

    cells = row.find_all('td')

    # Skip the header row
    if not cells:
        continue


    # --------------------------------------------------
    # 5. Extract product name
    # Using get_text()
    # --------------------------------------------------

    product_name = cells[0].get_text(strip=True)


    # --------------------------------------------------
    # 6. Extract description
    # Using get_text()
    # --------------------------------------------------

    description = cells[1].get_text(strip=True)


    # --------------------------------------------------
    # 7. Extract price
    # Using get_text()
    # --------------------------------------------------

    price = cells[2].get_text(strip=True)


    # --------------------------------------------------
    # 8. Extract image
    # --------------------------------------------------

    image_tag = cells[3].find('img')

    if image_tag:
        image_url = image_tag.get('src')
    else:
        image_url = None


    # --------------------------------------------------
    # 9. Print product information
    # --------------------------------------------------

    print('=' * 60)

    print(f'Product Name : {product_name}')
    print(f'Description  : {description}')
    print(f'Price        : {price}')
    print(f'Image URL    : {image_url}')

    print('=' * 60)
```
